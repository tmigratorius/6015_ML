## Encode Sequence Data in Python

Machine learning algorithms cannot work with categorical data directly.

Categorical data must be converted to numbers.


[Useful Link](https://machinelearningmastery.com/how-to-one-hot-encode-sequence-data-in-python/)

 ### <font color='red'>transfer catagories into numbers, nn and deep learning cant use class, must use nums. 2 types of encoding</font>
 
  ### <font color='red'>1. assign binary nums </font>
 
  ### <font color='red'>2. for each cat, each row has col, put 0 on all non-applicable, 1 on applicable (1-hot)</font>

**What is One Hot Encoding?**

A one hot encoding is a representation of categorical variables as binary vectors.

This first requires that the categorical values be mapped to integer values.

Then, each integer value is represented as a binary vector that is all zero values except the index of the integer, which is marked with a 1.

**Example**

Given the data: 'red', 'red', 'green', apply encoding mechanisms:      
1- integer: 0, 0, 1        
2- One hot coding:            
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;\[1,0\],             
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;\[1,0\],              
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;\[0,1\],            



We could use an integer encoding directly, rescaled where needed. This may work for problems where there is a natural ordinal relationship between the categories.

<font color='red'>practice these</font>


## One Hot Encoding (from scratch)

```python
from numpy import argmax
# define input string
data = 'hello world'
print(data)
# define universe of possible input values
alphabet = 'abcdefghijklmnopqrstuvwxyz '
# define a mapping of chars to integers
char_to_int = dict((c, i) for i, c in enumerate(alphabet))
int_to_char = dict((i, c) for i, c in enumerate(alphabet))
# integer encode input data
integer_encoded = [char_to_int[char] for char in data]
print(integer_encoded)
# one hot encode
onehot_encoded = list()
for value in integer_encoded:
    letter = [0 for _ in range(len(alphabet))]
    letter[value] = 1
    onehot_encoded.append(letter)
print(onehot_encoded)
# invert encoding
inverted = int_to_char[argmax(onehot_encoded[0])]
print(inverted)

print(len(onehot_encoded),len(integer_encoded))
```

## One Hot Encoding (scikit-learn)

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
# define example
data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
values = array(data)
print(values)
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
# invert first example
inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
print(inverted)
```

## One Hot Encoding (Keras)

```python
from numpy import array
from numpy import argmax
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

# define example
data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
values = array(data)
print(values)
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)

# one hot encode
encoded = to_categorical(integer_encoded)
print(encoded)
# invert encoding
inverted = argmax(encoded[0])
print(inverted)
```

# MNIST Dataset(https://keras.io/datasets/)

## Multi-class problem

Dataset of 60,000 28x28 grayscale images of the 10 digits, along with a test set of 10,000 images.



**Remember deep learning is not itself a classifier rather it is a feature learning approach. As such, you can use it to learn representation and you need to end up with a layer of classification (binary or mutli-class) or regression.  It is a feature representation to do a task**

**In spite of the fact it is a high-performing learning approach, but you need to use the best hyper-parameter combination to minimize the computational resourse you have**

```python
#The Mnist database (http://yann.lecun.com/exdb/mnist/)

import warnings
warnings.filterwarnings("ignore")

from keras.datasets import mnist


# Step 1: Load the data
(train_images, train_labels),(test_images,test_labels)=mnist.load_data()

#Step 2: See the data distribution
print(len(train_labels))
print(train_labels)
print(train_images.shape)
print(len(test_labels))
print(test_labels)
print(test_images.shape)
```

```python
#Step 3: Prpare the tensors and normalize and have them float

#28*28 = 784
# in the images, 0-255 scale, that's the intensity of the pixels. divide by 255, normalized the data
#matrix should be tensorflow dimensions. Loaded, it's not in the right dims, first dim is each element, 
##2and and 3rd are the size of the img, 28 x 28
train_images=train_images.reshape((60000,28*28))
train_images=train_images.astype('float32')/255
test_images=test_images.reshape((10000,28*28))
test_images=test_images.astype('float32')/255
```

```python
#Step 4: Encode the categories
#need to encode, to one hot vectors with the category
from keras.utils import to_categorical
train_labels=to_categorical(train_labels)
test_labels=to_categorical(test_labels)
```

```python

print(len(train_labels),train_labels)
```

```python
print(len(test_labels),test_labels)
```

## Activation Functions and Loss Functions in Last Layers

 
 

[Activation Function in Keras](https://keras.io/activations/)

<font color='red'>the interior activations are up to yoou, this is for final and loss only</font>

```python
#Build the model
from keras import models
from keras import layers
network =models.Sequential()
#there are two layes here, first is 512 because it's 2^9
network.add(layers.Dense(512,activation='relu',input_shape=(28*28,)))
#this layer is output layer, 10 categories, softmax becaure its a multiclass, single-lable classification
network.add(layers.Dense(10,activation='softmax'))
network.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])

#Relu: https://en.wikipedia.org/wiki/Rectifier_(neural_networks)
```

## Question (Interpret the above model with the summary)?

```python
network.summary()
#28*28*512+512=401921
#why add 512? this is the bias for every node.

#10*512+10 = 5130
# this is from the previous node, 512+10 10 is the bias
```

```python
network.fit(train_images,train_labels,epochs=5,batch_size=100)
```

```python
test_loss,test_acc=network.evaluate(test_images,test_labels)
print('test_acc',test_acc)

#Overfitting?
#nns hard because there are lots of hyperpramaters, hard to tune
#last layer is what your'e interested in, clustering, whatever. each layer tries to extract info from the previous layer
```

# [IMDB Database ](https://www.imdb.com/interfaces/)

## Binary class problem

The IMDB dataset: a set of 50,000 highly polarized reviews from the Internet Movie Database. They’re split into 25,000 reviews for training and 25,000 reviews for testing, each set consisting of 50% negative and 50% positive reviews.

```python
#Step 1: Load the data
from keras.datasets import imdb
(train_data,train_labels),(test_data,test_labels)=imdb.load_data(num_words=10000)

#The argument num_words=10000 means you’ll only keep the top 10,000 most frequently occurring words in the training data.
```

```python
#See the data record
print(train_data[0])
print(len(train_data[0]))
```

```python
#See the data record
print(train_data[10])
print(len(train_data[10]))
```

```python
#From previous two records, we need:
#1- Unify the length of feature vector
#2- Do one-hot encoder

import numpy as np
def vactorize_sequences(sequences,dimension=10000):
    results= np.zeros((len(sequences),dimension))
    for i,sequence in enumerate(sequences):
        results[i,sequence] = 1
    return results
x_train=vactorize_sequences(train_data)
x_test=vactorize_sequences(test_data)
```

```python
print(list(x_train[0]))
```

```python
# float the labels
#***always do this
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

```python
#Let us take part of array as a validation dataset
x_val=x_train[:10000]
partial_x_train=x_train[10000:]
y_val=y_train[:10000]
partial_y_train=y_train[10000:]
```

```python
#Build the model, compile. and validate using validation dataset
from keras import models
from keras import layers
model= models.Sequential()
#*** Input layer is first, shape to make tensor with 10k
#*** 16 is 2^someting
model.add(layers.Dense(16, activation='relu',input_shape=(10000,)))
#**2nd takes output from first
model.add(layers.Dense(16, activation='relu'))
#** why this one 1: output is binary, really only 1 lable
model.add(layers.Dense(1, activation='sigmoid'))
#*** what is compile: specifies the optimizer, loss fn, and what metrics used (accuracy), Puts the whole model together
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['acc'])

#Build history at each epoch
#***this is training
history=model.fit(partial_x_train,
                 partial_y_train,
                 epochs=20,
                 batch_size=512,
                 validation_data=(x_val,y_val))
```

```python
model.summary()
#***must know where to get paramters andwhat they mean
```

### Do we have overfitting?

```python
import matplotlib.pyplot as plt
history_dict=history.history
loss_values=history_dict['loss']
val_loss_values=history_dict['val_loss']
epochs=range(1,len(loss_values)+1)
plt.plot(epochs,loss_values,'bo',label='Training loss')
plt.plot(epochs,val_loss_values,'b',label='Validation loss')
plt.title("Training and validation loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

#*** see if there is overfitting, diverging training and validation loss is overfit
```

```python
acc_values=history_dict['acc']
val_acc_values=history_dict['val_acc']
plt.plot(epochs,acc_values,'bo',label='Training acc')
plt.plot(epochs,val_acc_values,'b',label='Validation acc')
plt.title("Training and validation acc")
plt.xlabel('Epochs')
plt.ylabel('accutacy')
plt.legend()
plt.show()

#*** similar behavior, training and validation accuracy diverge after so many epochs, overfitting

#*** how to get the best value of epochs before overfitting starts, 3-ish looks good
```

The training loss decreases with every epoch, and the training accuracy increases with every epoch. That’s what you would expect when running gradient descent optimization.

We see the overfitting and also maybe the best choice for epoch = 3

```python
#Build the model, compile. and validate using validation dataset

#*** redo the model with 3, check accuracy againg
from keras import models
from keras import layers
model= models.Sequential()
model.add(layers.Dense(16, activation='relu',input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['acc'])


history=model.fit(partial_x_train,
                 partial_y_train,
                 epochs=3,
                 batch_size=512,
                 validation_data=(x_val,y_val))
```

```python
import matplotlib.pyplot as plt
history_dict=history.history
loss_values=history_dict['loss']
val_loss_values=history_dict['val_loss']
epochs=range(1,len(loss_values)+1)
plt.plot(epochs,loss_values,'bo',label='Training loss')
plt.plot(epochs,val_loss_values,'b',label='Validation loss')
plt.title("Training and validation loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#Build the model, compile. and validate using validation dataset
#*** rebuild model again with ALL data, put val data back in. Train the best model on the whole dataset

from keras import models
from keras import layers
model= models.Sequential()
model.add(layers.Dense(16, activation='relu',input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['acc'])


history=model.fit(x_train,
                 y_train,
                 epochs=3,
                 batch_size=512,
                 validation_data=(x_val,y_val))
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
print('test_acc',test_acc)
```
